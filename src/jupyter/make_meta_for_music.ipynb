```python
start = 1711542997.2122576
df['start_delta'] = df['Song_start_time'] - start
df['end_delta'] = df['Song_end_time'] - start
df['start_index'] = (df['start_delta'] // 0.004).astype(int)
df['end_index'] = (df['end_delta'] // 0.004).astype(int)
df
```

```python
song_to_id = {song: ind for ind, song in enumerate(sorted(df['Theme']))}
print(len(song_to_id))
```

```python
import torch 

song_ids = torch.ones(390_000, dtype=int) * -1
is_liked = torch.ones(390_000, dtype=int) * -1
for _, row in df.iterrows():
    start_ind = row['start_index']
    end_ind = row['end_index']
    song_ids[start_ind : end_ind] = song_to_id[row['Theme']]
    is_liked[start_ind : end_ind] = 1 if row['Valence'] >= 3 else 0
```
